**Word Embedding by using Keras**

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
### sentences
sent=[  'that so kind of you',
     'please help yourself',
     'how dare you',
    'sorry for the inconvenience',
     'would you please speak slowly',
     'how are you',
     'i am so happy',]
```

```python
sent
```

```python
voc_size=10000   # We can declare vocabulary size of any.
```

```python
onehot_repr=[one_hot(words,voc_size)for words in sent] 
print(onehot_repr) # printing one hot representation 
```

```python
sent_length=8   # Here we fixed sentence length 
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length) # If sentence is not having 8 words then it add/pad 0, that is also beginning
print(embedded_docs)
```

```python
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse') # Mean Squared Error (MSE) is perhaps the simplest and most common loss function
```

```python
model.summary() # Printing model summary
```

```python
print(model.predict(embedded_docs))
```
